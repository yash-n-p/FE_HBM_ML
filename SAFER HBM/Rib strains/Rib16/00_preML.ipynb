# Initial Data

```python
plt.figure(figsize=(10,7))
sns.distplot(y,hist=True)
```

# SQRT

```python
y_2 = np.sqrt(y)
plt.figure(figsize=(10,7))
sns.distplot(y_2,hist=True)
y_2.skew()
```

# Cube root

```python
y_3 = np.cbrt(y)
plt.figure(figsize=(10,7))
sns.distplot(y_3,hist=True)
y_3.skew()
```

# Logp

```python
y_lg = np.log(y)
plt.figure(figsize=(10,7))
sns.distplot(y_lg,hist=True)
y_lg.skew()
```
